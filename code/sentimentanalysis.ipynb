```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from transformers import BertTokenizer, BertForSequenceClassification
import torch
from torch.nn.functional import softmax

# Ensure NLTK resources are downloaded
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

class TextPreprocessor:
    def __init__(self, model_name='bert-base-multilingual-cased'):
        self.lemmatizer = WordNetLemmatizer()
        # English stopwords from NLTK
        english_stop_words = set(stopwords.words('english'))
        # Swahili stopwords
        swahili_stop_words = set(['na', 'ya', 'ni', 'kwa', 'wa'])
        # Combine English and Swahili stopwords
        self.stop_words = english_stop_words.union(swahili_stop_words)
        self.tokenizer = BertTokenizer.from_pretrained(model_name)
        self.model = BertForSequenceClassification.from_pretrained(model_name)
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.model.to(self.device)

    def clean_and_preprocess(self, text):
        # Normalize the text to lowercase
        text = text.lower()
        # Tokenize the text
        tokens = word_tokenize(text)
        # Remove stopwords
        tokens = [word for word in tokens if word not in self.stop_words]
        # Lemmatize the remaining tokens
        lemmatized_text = ' '.join([self.lemmatizer.lemmatize(word) for word in tokens])
        return lemmatized_text

    def get_sentiment_score(self, text):
        # Prepare the text for BERT model input
        inputs = self.tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512).to(self.device)
        # Predict sentiment with the model
        with torch.no_grad():
            outputs = self.model(**inputs)
            predictions = softmax(outputs.logits, dim=1)
        # Dynamically extract sentiment scores based on the number of output classes
        scores = predictions[0].tolist()  # Convert the first (and only) batch of predictions to a list
        if len(scores) == 3:  # Expected number of classes for negative, neutral, and positive
            negative_score, neutral_score, positive_score = scores
        elif len(scores) == 2:  # Fallback for binary classification (e.g., positive and negative)
            negative_score, positive_score = scores
            neutral_score = None  # Neutral score is not applicable
        else:
            raise ValueError("Unexpected number of output classes: {}".format(len(scores)))
        return negative_score, neutral_score, positive_score

   

# Initialize the TextPreprocessor
preprocessor = TextPreprocessor()

# Apply cleaning and preprocessing to the text column
data['cleaned_text'] = data['text'].apply(lambda x: preprocessor.clean_and_preprocess(x) if pd.notnull(x) else '')

# Generate sentiment scores for the cleaned text
data['sentiment_score'] = data['cleaned_text'].apply(lambda x: preprocessor.get_sentiment_score(x) if x != '' else None)

# Display the updated DataFrame with sentiment scores
data[['text', 'cleaned_text', 'sentiment_score']]
```

```python
class EABLSentimentAnalyzerRefactored:
    def __init__(self, data):
        self.data = data

    def classify_sentiments(self, threshold_positive=0.5, threshold_negative=-0.5):
        def classify(score):
            if score > threshold_positive:
                return 'positive'
            elif score < threshold_negative:
                return 'negative'
            else:
                return 'neutral'
        self.data['sentiment_category'] = self.data['sentiment_score'].apply(classify)

    def sentiment_over_time(self, period='M'):
        # Make sure to use the correct column name for your date. Adjust 'Date' if necessary to match your DataFrame's column.
        self.data['date'] = pd.to_datetime(self.data['Date'])  # Corrected to 'date'
        sentiment_over_time_df = self.data.resample(period, on='date').mean()['sentiment_score'].reset_index()  # Use 'date'
        return sentiment_over_time_df

    def quantify_sentiment_shifts(self):
        sentiment_over_time = self.sentiment_over_time('D')['sentiment_score']
        sentiment_shifts_df = sentiment_over_time.diff().dropna().reset_index()
        sentiment_shifts_df.columns = ['date', 'sentiment_shift']  # Corrected to 'date'
        return sentiment_shifts_df

    def sentiment_distribution(self):
        sentiment_distribution_df = self.data['sentiment_category'].value_counts().reset_index()
        sentiment_distribution_df.columns = ['sentiment_category', 'count']
        return sentiment_distribution_df

    def top_tweets(self, sentiment, n=5):
        filtered_tweets_df = self.data[self.data['sentiment_category'] == sentiment]
        top_tweets_df = filtered_tweets_df.nlargest(n, 'sentiment_score')[['cleaned_text', 'sentiment_score']]  # Adjusted to 'text'
        return top_tweets_df

# Instantiate and use the refactored EABLSentimentAnalyzer
analyzer_refactored = EABLSentimentAnalyzerRefactored(data)

# Classify sentiments
analyzer_refactored.classify_sentiments()

# Analyze sentiment over time as DataFrame
sentiment_over_time_df = analyzer_refactored.sentiment_over_time(period='D')

# Quantify sentiment shifts as DataFrame
sentiment_shifts_df = analyzer_refactored.quantify_sentiment_shifts()

# Sentiment distribution as DataFrame
sentiment_distribution_df = analyzer_refactored.sentiment_distribution()

# Top positive and negative tweets as DataFrame
top_positive_tweet_df = analyzer_refactored.top_tweets('positive', n=3)
top_neutral_tweet_df = analyzer_refactored.top_tweets('neurtal', n=3)

```

# ***2.3.CORRELATION ANALYSIS*** 

• Explore the relationship between product sentiment scores and stock market performance using 
statistical techniques.

• Identify correlations and causal relationships between changes in sentiment and fluctuations in 
stock prices.

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr, spearmanr
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

class CorrelationAnalysis:
    def __init__(self, data):
        self.data = data

    def calculate_correlation(self, method='pearson'):
        if method == 'pearson':
            corr, p_value = pearsonr(self.data['sentiment_score'], self.data['Close'])
        elif method == 'spearman':
            corr, p_value = spearmanr(self.data['sentiment_score'], self.data['Close'])
        else:
            raise ValueError("Unsupported correlation method. Use 'pearson' or 'spearman'.")
        return corr, p_value

    def perform_regression_analysis(self):
        X = sm.add_constant(self.data['sentiment_score'])
        y = self.data['Close']
        model = sm.OLS(y, X).fit()
        return model.summary()

    def visualize_trends(self, window_size=50):
        """
        Visualizes the trends of sentiment scores and stock closing prices over time, with smoothed sentiment scores.
        
        Parameters:
        window_size (int): The window size for calculating the rolling mean of the sentiment scores.
        """
        # Ensure the 'date' column is in datetime format
        self.data['date'] = pd.to_datetime(self.data['Date'])

        # Calculate rolling mean of sentiment scores for smoothing
        self.data['sentiment_score_smoothed'] = self.data['sentiment_score'].rolling(window=window_size).mean()

        # Create the plot
        fig, ax1 = plt.subplots(figsize=(8, 6))

        # Plot sentiment scores
        color = 'tab:red'
        ax1.set_xlabel('Date')
        ax1.set_ylabel('Sentiment Score', color=color)
        ax1.plot(self.data['date'], self.data['sentiment_score_smoothed'], color=color, label='Sentiment Score (Smoothed)')
        ax1.tick_params(axis='y', labelcolor=color)
        ax1.legend(loc='upper left')

        # Plot stock closing prices
        ax2 = ax1.twinx()
        color = 'tab:blue'
        ax2.set_ylabel('Stock Closing Price', color=color)
        ax2.plot(self.data['date'], self.data['Close'], color=color, label='Stock Closing Price')
        ax2.tick_params(axis='y', labelcolor=color)
        ax2.legend(loc='upper right')

        # Rotate date labels for better readability
        plt.xticks(rotation=45)
        
        # Title and layout adjustments
        plt.title('Sentiment Score and Stock Closing Price Trends')
        fig.tight_layout()

        # Show the plot
        plt.show()

analysis = CorrelationAnalysis(data)
analysis.visualize_trends()
analysis.perform_regression_analysis()
```



## ***Visualization Observations:***

- The visualization includes dates on the x-axis, ranging from 2014 to 2024, providing a temporal context to the sentiment and stock price trends.

- The sentiment score (smoothed) and the stock closing price both show fluctuations over time.

- There are moments when the trends in sentiment scores and stock prices appear to move in tandem, and at other times they diverge.

## ***Regression Analysis Observations:***

- **R-squared value (0.003)**: This value indicates that the model explains a very small portion of the variance in stock prices based on sentiment scores.

- **F-statistic (3.560)**: The the corresponding p-value is closer to the conventional significance level (0.0595), suggesting that the model could be nearing a point of statistical significance.

- **Coefficients**:
 
  - The intercept (`const`) is highly significant (p-value of 0.000), which means the model predicts a substantial baseline stock price regardless of sentiment.
  
  - The coefficient for `sentiment_score` is negative and approaching significance (p-value of 0.059), indicating that an inverse relationship might exist between sentiment scores and stock prices, albeit this relationship is not strong.

- **Durbin-Watson statistic (0.015)**: indicates a strong positive autocorrelation in the residuals, which means consecutive observations are not independent. This could be problematic for the regression model as it violates one of the OLS assumptions.

- **Normality Tests (Omnibus, Jarque-Bera)**: Both tests suggest that the residuals of the model do not follow a normal distribution, which is another violation of OLS assumptions.

## ***Interpretations:***

- The  R-squared value and the  F-statistic suggest that there might be a weak relationship between sentiment scores and stock prices that was not captured in the initial model. However, this relationship is not strong enough to make reliable predictions.

- The negative coefficient for sentiment scores implies that higher sentiment scores could be associated with lower stock prices, or vice versa, but this relationship is not statistically robust at this point.

## ***Insights:***

- The lack of a strong statistical relationship in the model and the OLS assumption violations suggest that a linear regression model may not be the best approach for this kind of analysis. Time-series data often have autocorrelation and non-normality that need to be accounted for with more specialized models.

- Considering that the data spans multiple years, external factors not accounted for in the model, such as market trends, economic events, and company performance, are likely influencing stock prices.

- The slight negative relationship indicated by the model suggests that sentiment may not be a leading indicator of stock performance in this case, or that the relationship is more complex than a simple linear model can capture.

- The visualization and model together suggest that while sentiment may have some impact on stock prices, it is not a dominant factor and is likely overshadowed by other market forces.

- Future analyses could benefit from incorporating additional variables, using different modeling techniques (e.g., ARIMA, VAR), or considering non-linear models to better capture the dynamics between sentiment and stock prices.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import spearmanr

class CorrelationAnalysis:
    def __init__(self, data):
        """
        Initialize with a pandas DataFrame that includes both sentiment scores and stock market indicators.
        :param data: pandas DataFrame. Must include columns for sentiment scores and at least one stock market indicator.
        """
        self.data = data

    def calculate_correlation(self, method='pearson'):
        """
        Calculate the correlation matrix using either Pearson or Spearman correlation coefficient.
        :param method: String, 'pearson' or 'spearman' for the type of correlation coefficient.
        :return: pandas DataFrame, correlation matrix.
        """
        if method not in ['pearson', 'spearman']:
            raise ValueError("Method must be either 'pearson' or 'spearman'.")
        
        if method == 'spearman':
            # Calculate Spearman correlation
            correlation_matrix = self.data.corr(method='spearman')
        else:
            # Calculate Pearson correlation by default
            correlation_matrix = self.data.corr(method='pearson')
        
        return correlation_matrix

    def visualize_correlation_matrix(self, method='pearson'):
        """
        Visualize the correlation matrix using a heatmap.
        :param method: String, 'pearson' or 'spearman' for the type of correlation coefficient.
        """
        correlation_matrix = self.calculate_correlation(method)
        plt.figure(figsize=(10, 8))
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
        plt.title(f"{method.capitalize()} Correlation Matrix")
        plt.show()

    def scatter_plot(self, column_x, column_y):
        """
        Generate a scatter plot for two specified columns from the DataFrame.
        :param column_x: String, column name for x-axis.
        :param column_y: String, column name for y-axis.
        """
        plt.figure(figsize=(8, 6))
        sns.scatterplot(data=self.data, x=column_x, y=column_y)
        plt.title(f"Scatter Plot of {column_x} vs {column_y}")
        plt.xlabel(column_x)
        plt.ylabel(column_y)
        plt.show()
    
    def calculate_spearman_rank_correlation(self, column_x, column_y):
        """
        Calculate Spearman's rank correlation coefficient for two specified columns.
        :param column_x: String, column name for the first variable.
        :param column_y: String, column name for the second variable.
        :return: Spearman's rank correlation coefficient and p-value.
        """
        coefficient, p_value = spearmanr(self.data[column_x], self.data[column_y])
        return coefficient, p_value
```

```python
# Assuming the CorrelationAnalysis class definition is already executed in your environment

# Initialize the CorrelationAnalysis with the sample data
correlation_analysis = CorrelationAnalysis(data)

# Calculate and print Pearson correlation matrix
pearson_corr = correlation_analysis.calculate_correlation(method='pearson')
print("Pearson Correlation Matrix:\n", pearson_corr)

# Calculate and print Spearman correlation matrix
spearman_corr = correlation_analysis.calculate_correlation(method='spearman')
print("\nSpearman Correlation Matrix:\n", spearman_corr)

# Visualize the Pearson correlation matrix
correlation_analysis.visualize_correlation_matrix(method='pearson')

# Visualize the Spearman correlation matrix
correlation_analysis.visualize_correlation_matrix(method='spearman')

# Generate a scatter plot between 'sentiment_score' and 'daily_close'
#correlation_analysis.scatter_plot('sentiment_score', 'Close')

# Calculate and print Spearman's rank correlation coefficient between 'sentiment_score' and 'daily_close'
spearman_rank_coefficient, p_value = correlation_analysis.calculate_spearman_rank_correlation('sentiment_score', 'Close')
print(f"\nSpearman's Rank Correlation Coefficient between 'sentiment_score' and 'daily_close': {spearman_rank_coefficient}, p-value: {p_value}")
```

## ***Insights:***

- The weak correlations suggest that there is not a strong direct relationship between sentiment scores and closing prices in the data. This could imply that other factors may have a more significant impact on stock market performance than sentiment scores alone.

- The smoothed sentiment score appears to have a more substantial relationship with closing prices than the raw sentiment score. This could suggest that stock prices may be more sensitive to the general trend or smoothed sentiment rather than daily fluctuations.

- The fact that Spearman's correlation is more negative than Pearson's when comparing 'Close' and 'sentiment_score_smoothed' suggests that the relationship may be more about the order or rank rather than the raw values. This might be important for investment strategies that rely on relative changes rather than absolute sentiment scores.

- It's essential to note that correlation does not imply causation. Therefore, even though there are statistically significant relationships, further analysis would be needed to determine if sentiment scores directly influence stock market performance or if they are both influenced by another factor.
